```python
import pandas as pd

em_data_filepath='/home/data_folder/uga/Impact_flood_drought_Uganda_UNDesinventor.xlsx'

dfb = pd.read_excel(em_data_filepath, engine='openpyxl',skiprows=0,sheet_name = 'UGA_Flood_and_Drought')
```

```python
#dfb['date']=pd.to_datetime(dfb['Date (YMD)'], format='%Y/%m/%-d')

dfb[['year','month','day']]= dfb['Date (YMD)'].str.split('/', 2, expand=True)

dfb["day1"] = dfb["day"].astype(int) + 1

dfb['month1'] = dfb['month'].replace('0', '1')
```

```python
dfb['year'].max()
```

```python
dfb['month2']=dfb.month1.astype(str).str.pad(2,fillchar='0')
dfb['day2']=dfb.day1.astype(str).str.pad(2,fillchar='0')

dfb["date"] = dfb["year"].astype(str) + dfb["month2"]+ dfb["day2"]

dfb['date1']=pd.to_datetime(dfb['date'], format='%Y%m%d')
```
